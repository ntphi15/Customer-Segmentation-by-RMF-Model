```python
# Read "customer_registered" table from "customer_registered.csv" file
import pandas as pd

customer_registered = pd.read_csv("C:\\Users\\Admin\\OneDrive - VNU-HCMUS\Data Analytics\\1. Customer Segmentation by RFM Model\\0. Raw Data\\customer_registered.csv")
customer_registered
```

```python
# Convert "created_date" column to Date type by using the Pandas Python module
customer_registered['created_date'] = pd.to_datetime(customer_registered['created_date']).dt.date
customer_registered['created_date'] = pd.to_datetime(customer_registered['created_date'])

# Convert "stopdate" column to Date type by using the Pandas Python module
customer_registered['stopdate'] = pd.to_datetime(customer_registered['stopdate']).dt.date
customer_registered['stopdate'] = pd.to_datetime(customer_registered['stopdate'])
customer_registered.info()
```
